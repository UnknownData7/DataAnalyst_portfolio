```python
from itertools import product

def solve_task():
    m, n, k = 7, 5, 3
    
    def generate_valid_rows(length):
        valid_rows = []
        for pattern in product(['.', 'ч'], repeat=length):
            white_count = 0
            valid = True
            for cell in pattern:
                if cell == '.':
                    white_count += 1
                    if white_count >= k:
                        valid = False
                        break
                else:
                    white_count = 0
            if valid:
                valid_rows.append(pattern)
        return valid_rows
    
    valid_rows = generate_valid_rows(n)
    
    total_cells = m * n
    max_black = total_cells // 3
    if total_cells % 3 == 0:
        max_black -= 1
    
    solutions = []
    solutions_count = 0
    
    for grid_rows in product(valid_rows, repeat=m):
        valid_grid = True
        for col in range(n):
            white_streak = 0
            for row in range(m):
                if grid_rows[row][col] == '.':
                    white_streak += 1
                    if white_streak >= k:
                        valid_grid = False
                        break
                else:
                    white_streak = 0
            if not valid_grid:
                break
        
        if not valid_grid:
            continue
        
        black_count = sum(row.count('ч') for row in grid_rows)
        if black_count <= max_black:
            solutions.append(grid_rows)
            solutions_count += 1
    
    for solution in solutions:
        for row in solution:
            print(''.join(row))
        print()
    
    print(f"Всего решений: {solutions_count}")
    return solutions_count

if __name__ == "__main__":
    result = solve_task()
```
