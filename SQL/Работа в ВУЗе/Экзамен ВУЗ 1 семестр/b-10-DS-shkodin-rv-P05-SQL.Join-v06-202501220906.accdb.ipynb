```python
import pyodbc
import pandas as pd
import numpy as np
import math
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Путь к БД:
database_path = r'C:\Users\Dn103-User07.STUD\Desktop\Start\database.accdb'

# Настройка подключения через ODBC драйвер
conn_str = (
    r'DRIVER={Microsoft Access Driver (*.mdb, *.accdb)};'
    rf'DBQ={database_path};'
)

try:
    #Создаем соединение
    conn = pyodbc.connect(conn_str)
    print('Подключение установленно успешно')

except pyodbc.Error as e:
     print("Ошибка при подключении к базе данных или выполнении запроса:", e)
```

ЗАДАНИЕ №56 CREATE TABLE


```python
try:
    # Подключение к базе данных
    conn = pyodbc.connect(conn_str)
    cursor = conn.cursor()

    # Выполнение SQL-запроса создание таблицы
    query = """
    CREATE TABLE tbUsers2 (
    UserID AUTOINCREMENT PRIMARY KEY, 
    FirstName VARCHAR(20),
    LastName VARCHAR(20),
    Email VARCHAR(30) UNIQUE);
    """
    # Выполнение SQL запроса 
    cursor.execute(query)
    conn.commit()  
    # Фиксируем изменения в базе данных
    print("Таблица tbUsers2 успешно создана и заполнена.")


except pyodbc.Error as db_err:
    print(f"Ошибка при выполнении SQL-запроса: {db_err}")
except ValueError as ve:
    print(f"Ошибка: {ve}")
except Exception as e:
    print(f"Произошла ошибка: {e}")
finally:
    if cursor: 
    # Закрытие соединения
        cursor.close()
    if conn: 
        conn.close()
```
